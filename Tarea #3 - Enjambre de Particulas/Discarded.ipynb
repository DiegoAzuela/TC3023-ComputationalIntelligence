Particle Swarm Optimization (PSO)
* gx = global best
* xix = local best
Other Variables
* xit = current x(t)
* xit_1 = xi(t+1)
* vit = current v(t)
* vit_1 = vi(t+1)

```python
def velocity_particle(nv,vit,xit,gx,xix):
  """
  Function: Computes the velocity of the particle based on:
    * nv, vit, xit, gx, xix, Epsilon, Beta
  Input: NA
  Output: The velocity of each particle is returned
  """
  list_vit = [] #List to store velocities
  #vit = 0  #Initial velocity
  for particle in range(nv):
    alpha = beta = 2
    epsilon_1 = np.random.rand() #vectores aleatorios con valores entre 0 y 1, size:tbd
    epsilon_2 = np.random.rand() #vectores aleatorios con valores entre 0 y 1, size:tbd
    #Particle velocity
    vit_1 = vit + (alpha*epsilon_1*(gx-xit)) + beta*epsilon_2*(xix-xit)
    #Enforce boundaries 
    if vit_1 > Vmax:
      vit_1 = Vmax
    list_vit.append(vit_1)
    vit = vit_1
    
  return vit_1
```

```python
def new_position(vit_1,xit):
  """
  Function: Computes the position of the particle based on:
    * New Velocity
    * Old Position
  Input: Enter the new velocity and the current position
  Output: Function evaluates new position
  """
  list_position = []
  #xit = 0 #needed??
  for particle in range(nv):
    xit_1 = xit + vit_1
    list_position.append(xit_1)
  return xit_1
```
