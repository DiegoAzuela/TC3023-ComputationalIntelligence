```
"""
Autor: Gabriel González Sahagún
Clase: Inteligencia Computacional TC3023
Descripción: Código para optimizar el problema del vendedor viajero (TSP) utilizando optimización de colonia
de hormigas (ACO).
"""

import numpy as np
from scipy.spatial import distance_matrix
import matplotlib.pyplot as plt
from multiprocessing.dummy import Pool as ThreadPool
from functools import partial
import math

# Definir el número de ciudades a visitar.
n_cities = 3
# Definir el número de iteraciones del algoritmo.
n_it = 10
# Definir el número de hormigas.
n_hormigas = 5
# Definir alpha, puede ser un valor de entre 0 y 1
alpha = 1 
# Definir beta, puede ser un valor de entre 0 y 1
beta = 1
# Definir Q
Q = 1000
# Definir rho
rho = 0.5

# Generar las coordenadas para n ciudades.
cities = np.random.randint(1, 100, size=(n_cities, 2))
# Calcular la matriz de distancias
dm = distance_matrix(cities, cities)
# Inicializar en 1 el valor de Tau para cada conexión.
tau = np.ones((n_cities, n_cities))
# Inicializar la matriz Eta con el inverso de la distancia entre cada par de ciudades (1/dij).
eta = np.asarray([1 / dm[i, j] for i in range(n_cities) for j in range(n_cities)]).reshape((n_cities, n_cities))
# Reemplazar el inifinito en la diagonal causado a dividir 1 entre 0 (la distancia a sí mismo).
eta = np.where(eta == np.Infinity, 0, eta)

mejor_sol = None
mejor_dist = np.Inf


def distancia(solucion):
    """
    Función que calcula la distancia recorrida de la hormiga.
    :param solución: lista que contiene la ruta de la hormiga.
    :return: distancia de la ruta.
    """
    conexiones = list(zip(solucion, solucion[1:] + [solucion[0]]))
    L = 0
    for i, j in conexiones:
        L += dm[i, j]
    return L

def tij(solucion, Q):
    """
    Función que calcula la cantidad de feromonas que deja la hormiga.
    :param solucion:
    :param Q: aportación de feromonas.
    :return: cantidad de feromonas que deja la hormiga.
    """
    return Q / distancia(solucion)

def correr_hormiga(ciudad, tau, n, alpha, beta):
    """
    Función que genera la ruta que toma una hormiga a partir del rastro de feromonas que se tiene.
    :param ciudad: ciudad inicial de la hormiga.
    :return: lista que contiene la ruta de la hormiga.
    """
    # Crear la ruta de la hormiga.
    ruta = [ciudad]
    # Crear la lista de las ciudades que no han sido visitadas.
    posibles = list(range(n_cities))
    # Remover la ciudad inicial.
    posibles.remove(ciudad)
    # Crear string con el log para imprimir toda la información y evitar que se mezclen los prints por los threads.
    log_hormiga = "Inicial: {}\n".format(ciudad)
    while (posibles):
        log_hormiga += "Posibles: {}\n".format(posibles)
        # Crear el vector probabilidades para ir de la ciudad i a las ciudades posibles.
        p = np.zeros(n_cities)
        suma = 0
        # Calcular la importancia para cada ciudad.
        for j in posibles:
            p[j] = np.power(tau[min(ciudad, j), max(ciudad, j)], alpha) * np.power(n[min(ciudad, j), max(ciudad, j)], beta)
            suma += p[j]
        # Normalizar la importancia para obtener la probabilidad.
        p = p / suma
        log_hormiga += "{}\n".format(p)
        # Generar un número random para determinar cual ciudad escoger.
        r = np.random.rand()
        # Acumulador de la probabilidad.
        suma = 0
        for index, pij in enumerate(p.tolist()):
            # Aumentar el acumulador de probabilidad.
            suma += pij
            # Determinar si se escoge la ciudad.
            if r < suma:
                # Establecer la siguiente ciudad como la ubicación actual.
                ciudad = index
                # Agregar la ciudad actual a la ruta.
                ruta.append(ciudad)
                # Remover la ciudad actual de la lista de posibles ciudades a visitar.
                posibles.remove(ciudad)
                break

    print(log_hormiga)
    return ruta

def actualizar_tau(caminos, tau, Q):
    """
    Función que actualiza la matriz tau que contiene la cantidad de feromonas por arco en el grafo.
    :param caminos: lista que contiene las rutas (listas) de las hormigas.
    :param tau: matriz que contiene la cantidad de feromonas en cada arco.
    :param Q: aportación de feromonas de cada hormiga.
    :return:
    """
    # Definir una función con un parámetro fijo para el map.
    calcular_feromonas = partial(tij, Q=Q)
    # Calcular la cantidad de feromonas que deja cada hormiga.
    feromonas_hormigas = list(map(calcular_feromonas, caminos))
    # Conservar rho% de las feromonas de la iteracion pasada.
    tau = rho * tau
    # Agregar las feromonas de cada hormiga.
    for k, camino in enumerate(caminos):
        arcos = list(zip(camino, camino[1:] + [camino[0]]))
        for i, j in arcos:
            tau[min(i, j), max(i, j)] = tau[min(i, j), max(i, j)] + feromonas_hormigas[k]

    return tau

def plot_ants(soluciones, cities, mejor, ultimo_mejor):
    """
    Función que dibuja la mejor ruta hasta el momento y las rutas de la última iteración.
    :param soluciones: lista de rutas de la última iteración.
    :param cities: posición de las ciudades para graficar.
    :param mejor: la mejor ruta encontrada hasta el momento.
    :param ultimo_mejor: índice de la mejor ruta de la iteración.
    :return:
    """
    # Calcular numero de hormigas y agregar 1 por la mejor solución.
    n_soluciones = len(soluciones) + 1
    # Separar coordenadas para graficar.
    cx = cities[:, 0]
    cy = cities[:, 1]
    # Definir una figura de tamaño fijo.
    fig, axes = plt.subplots(math.ceil(n_soluciones/2), 2, figsize=(10,15))

    for i, ax in enumerate(axes.flatten()):
        # Dibujar los puntos de las ciudades.
        if i==0:
            # Dibujar ciudades.
            ax.scatter(cx, cy, c='r')
            # Agregar título.
            ax.set_title('Mejor solución')
            # Dibujar ruta.
            ax.plot(cx[mejor + [mejor[0]]], cy[mejor + [mejor[0]]], c='r')
            ax.set_aspect('auto')
        elif i==1:
            # Dibujar ciudades
            ax.scatter(cx, cy, c='r')
            # Agregar título.
            ax.set_title('Mejor de la iteración')
            # Sacar el mejor de la ruta con el índice.
            ant = soluciones.pop(ultimo_mejor)
            # Dibujar ruta.
            ax.plot(cx[ant + [ant[0]]], cy[ant + [ant[0]]], c='y')
            ax.set_aspect('auto')
        elif soluciones:
            # Dibujar ciudades
            ax.scatter(cx, cy, c='r')
            # Sacar el primero de la lista.
            ant = soluciones.pop(0)
            # Dibujar ruta.
            ax.plot(cx[ant + [ant[0]]], cy[ant + [ant[0]]], c='b')
            ax.set_aspect('auto')

    # Agregar espacios en blanco entre gráficas.
    plt.subplots_adjust(hspace=0.5)
    plt.show()

for contador in range(n_it):
    # Generar las ciudades iniciales para cada hormiga.
    ciudades_hormigas = np.random.randint(0, n_cities, n_hormigas)
    # Crear 4 threads.
    pool = ThreadPool(4)
    # Correr cada hormiga en un thread diferente.
    procesar_hormiga = partial(correr_hormiga, tau=tau, n=eta, alpha=alpha, beta=beta)
    resultados = list(pool.map(procesar_hormiga, ciudades_hormigas))

    # Cerrar los threads.
    pool.close()
    # Esperar a que los threads regresen el resultado.
    pool.join()
    # Actualizar el rastro de feromonas.
    tau = actualizar_tau(resultados, tau, Q)

    evaluaciones = np.asarray(list(map(lambda solucion: distancia(solucion), resultados)))
    print(evaluaciones)
    index = np.argmin(evaluaciones)
    dist = evaluaciones[index]
    print('TAU')
    np.savetxt("tau.csv", np.round(tau,2), delimiter=",")
    print(np.round(tau,3))
    print('ETA')
    np.savetxt("eta.csv", np.round(eta, 3), delimiter=",")
    print(np.round(eta, 3))
    print(resultados)
    print(np.round(evaluaciones, 3))
    if dist < mejor_dist:
        mejor_sol = resultados[index]
        mejor_dist = dist
    print('El mejor de la iteracion fue {} con {} m.'.format(resultados[index], dist))
    print('La mejor solucion es {} con {} m.'.format(mejor_sol, mejor_dist))
    plot_ants(soluciones=resultados, cities=cities, mejor=mejor_sol, ultimo_mejor=index)
```

```
"""
Autor: Gabriel González Sahagún
Clase: Inteligencia Computacional TC3023
Descripción: Código para optimizar el problema del vendedor viajero (TSP) utilizando optimización de colonia
de hormigas (ACO).
"""

import numpy as np
from scipy.spatial import distance_matrix
import matplotlib.pyplot as plt
from multiprocessing.dummy import Pool as ThreadPool
from functools import partial
import math

# Definir el número de ciudades a visitar.
n_cities = 5
# Definir el número de iteraciones del algoritmo.
n_it = 10
# Definir el número de hormigas.
n_hormigas = 5
# Definir alpha, puede ser un valor de entre 0 y 1
alpha = 1 
# Definir beta, puede ser un valor de entre 0 y 1
beta = 1
# Definir Q
Q = 1000
# Definir rho
rho = 0.5

# Generar las coordenadas para n ciudades.
cities = np.random.randint(1, 100, size=(n_cities, 2))
# Calcular la matriz de distancias
dm = distance_matrix(cities, cities)
# Inicializar en 1 el valor de Tau para cada conexión.
tau = np.ones((n_cities, n_cities))
# Inicializar la matriz Eta con el inverso de la distancia entre cada par de ciudades (1/dij).
eta = np.asarray([1 / dm[i, j] for i in range(n_cities) for j in range(n_cities)]).reshape((n_cities, n_cities))
# Reemplazar el inifinito en la diagonal causado a dividir 1 entre 0 (la distancia a sí mismo).
eta = np.where(eta == np.Infinity, 0, eta)

mejor_sol = None
mejor_dist = np.Inf


def distancia(solucion):
    """
    Función que calcula la distancia recorrida de la hormiga.
    :param solución: lista que contiene la ruta de la hormiga.
    :return: distancia de la ruta.
    """
    conexiones = list(zip(solucion, solucion[1:] + [solucion[0]]))
    L = 0
    for i, j in conexiones:
        L += dm[i, j]
    return L

def tij(solucion, Q):
    """
    Función que calcula la cantidad de feromonas que deja la hormiga.
    :param solucion:
    :param Q: aportación de feromonas.
    :return: cantidad de feromonas que deja la hormiga.
    """
    return Q / distancia(solucion)

def correr_hormiga(ciudad, tau, n, alpha, beta):
    """
    Función que genera la ruta que toma una hormiga a partir del rastro de feromonas que se tiene.
    :param ciudad: ciudad inicial de la hormiga.
    :return: lista que contiene la ruta de la hormiga.
    """
    # Crear la ruta de la hormiga.
    ruta = [ciudad]
    # Crear la lista de las ciudades que no han sido visitadas.
    posibles = list(range(n_cities))
    # Remover la ciudad inicial.
    posibles.remove(ciudad)
    # Crear string con el log para imprimir toda la información y evitar que se mezclen los prints por los threads.
    log_hormiga = "Inicial: {}\n".format(ciudad)
    while (posibles):
        log_hormiga += "Posibles: {}\n".format(posibles)
        # Crear el vector probabilidades para ir de la ciudad i a las ciudades posibles.
        p = np.zeros(n_cities)
        suma = 0
        # Calcular la importancia para cada ciudad.
        for j in posibles:
            p[j] = np.power(tau[min(ciudad, j), max(ciudad, j)], alpha) * np.power(n[min(ciudad, j), max(ciudad, j)], beta)
            suma += p[j]
        # Normalizar la importancia para obtener la probabilidad.
        p = p / suma
        log_hormiga += "{}\n".format(p)
        # Generar un número random para determinar cual ciudad escoger.
        r = np.random.rand()
        # Acumulador de la probabilidad.
        suma = 0
        for index, pij in enumerate(p.tolist()):
            # Aumentar el acumulador de probabilidad.
            suma += pij
            # Determinar si se escoge la ciudad.
            if r < suma:
                # Establecer la siguiente ciudad como la ubicación actual.
                ciudad = index
                # Agregar la ciudad actual a la ruta.
                ruta.append(ciudad)
                # Remover la ciudad actual de la lista de posibles ciudades a visitar.
                posibles.remove(ciudad)
                break

    print(log_hormiga)
    return ruta

def actualizar_tau(caminos, tau, Q):
    """
    Función que actualiza la matriz tau que contiene la cantidad de feromonas por arco en el grafo.
    :param caminos: lista que contiene las rutas (listas) de las hormigas.
    :param tau: matriz que contiene la cantidad de feromonas en cada arco.
    :param Q: aportación de feromonas de cada hormiga.
    :return:
    """
    # Definir una función con un parámetro fijo para el map.
    calcular_feromonas = partial(tij, Q=Q)
    # Calcular la cantidad de feromonas que deja cada hormiga.
    feromonas_hormigas = list(map(calcular_feromonas, caminos))
    # Conservar rho% de las feromonas de la iteracion pasada.
    tau = rho * tau
    # Agregar las feromonas de cada hormiga.
    for k, camino in enumerate(caminos):
        arcos = list(zip(camino, camino[1:] + [camino[0]]))
        for i, j in arcos:
            tau[min(i, j), max(i, j)] = tau[min(i, j), max(i, j)] + feromonas_hormigas[k]

    return tau

def plot_ants(soluciones, cities, mejor, ultimo_mejor):
    """
    Función que dibuja la mejor ruta hasta el momento y las rutas de la última iteración.
    :param soluciones: lista de rutas de la última iteración.
    :param cities: posición de las ciudades para graficar.
    :param mejor: la mejor ruta encontrada hasta el momento.
    :param ultimo_mejor: índice de la mejor ruta de la iteración.
    :return:
    """
    # Calcular numero de hormigas y agregar 1 por la mejor solución.
    n_soluciones = len(soluciones) + 1
    # Separar coordenadas para graficar.
    cx = cities[:, 0]
    cy = cities[:, 1]
    # Definir una figura de tamaño fijo.
    fig, axes = plt.subplots(math.ceil(n_soluciones/2), 2, figsize=(10,15))

    for i, ax in enumerate(axes.flatten()):
        # Dibujar los puntos de las ciudades.
        if i==0:
            # Dibujar ciudades.
            ax.scatter(cx, cy, c='r')
            # Agregar título.
            ax.set_title('Mejor solución')
            # Dibujar ruta.
            ax.plot(cx[mejor + [mejor[0]]], cy[mejor + [mejor[0]]], c='r')
            ax.set_aspect('auto')
        elif i==1:
            # Dibujar ciudades
            ax.scatter(cx, cy, c='r')
            # Agregar título.
            ax.set_title('Mejor de la iteración')
            # Sacar el mejor de la ruta con el índice.
            ant = soluciones.pop(ultimo_mejor)
            # Dibujar ruta.
            ax.plot(cx[ant + [ant[0]]], cy[ant + [ant[0]]], c='y')
            ax.set_aspect('auto')
        elif soluciones:
            # Dibujar ciudades
            ax.scatter(cx, cy, c='r')
            # Sacar el primero de la lista.
            ant = soluciones.pop(0)
            # Dibujar ruta.
            ax.plot(cx[ant + [ant[0]]], cy[ant + [ant[0]]], c='b')
            ax.set_aspect('auto')

    # Agregar espacios en blanco entre gráficas.
    plt.subplots_adjust(hspace=0.5)
    plt.show()

for contador in range(n_it):
    # Generar las ciudades iniciales para cada hormiga.
    ciudades_hormigas = np.random.randint(0, n_cities, n_hormigas)
    # Crear 4 threads.
    pool = ThreadPool(4)
    # Correr cada hormiga en un thread diferente.
    procesar_hormiga = partial(correr_hormiga, tau=tau, n=eta, alpha=alpha, beta=beta)
    resultados = list(pool.map(procesar_hormiga, ciudades_hormigas))

    # Cerrar los threads.
    pool.close()
    # Esperar a que los threads regresen el resultado.
    pool.join()
    # Actualizar el rastro de feromonas.
    tau = actualizar_tau(resultados, tau, Q)

    evaluaciones = np.asarray(list(map(lambda solucion: distancia(solucion), resultados)))
    print(evaluaciones)
    index = np.argmin(evaluaciones)
    dist = evaluaciones[index]
    print('TAU')
    np.savetxt("tau.csv", np.round(tau,2), delimiter=",")
    print(np.round(tau,3))
    print('ETA')
    np.savetxt("eta.csv", np.round(eta, 3), delimiter=",")
    print(np.round(eta, 3))
    print(resultados)
    print(np.round(evaluaciones, 3))
    if dist < mejor_dist:
        mejor_sol = resultados[index]
        mejor_dist = dist
    print('El mejor de la iteracion fue {} con {} m.'.format(resultados[index], dist))
    print('La mejor solucion es {} con {} m.'.format(mejor_sol, mejor_dist))
    plot_ants(soluciones=resultados, cities=cities, mejor=mejor_sol, ultimo_mejor=index)
```
